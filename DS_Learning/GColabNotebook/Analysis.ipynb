#####Sample Example/Coding for Overview of how to use "Normalization,Batchnormalization,Dropout,etc.." and Splitting of Train and Validation/Test Data and Applying "Optimizer,loss,metrics" before going into ML Automation Project.

```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
data
```

```python
x_columns = list(set(data.columns)-set({"median_house_value"}))
y_column = "median_house_value"
```

```python
x_data = data[x_columns].to_numpy()
y_data = data[y_column].to_numpy()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data,y_data,test_size=0.2)
```

```python
help(train_test_split)
```

```python
x_data.shape
```

```python
y_data.shape
```

```python
input_layer = tf.keras.layers.Input(shape=[8,])
normalization = tf.keras.layers.Normalization()(input_layer)
dense1 = tf.keras.layers.Dense(512,activation = "relu")(normalization)
dropout = tf.keras.layers.Dropout(0.2)(dense1)
dense2 = tf.keras.layers.Dense(256,activation="relu")(dropout)
batchnorm = tf.keras.layers.BatchNormalization()(dense2)
dense3 = tf.keras.layers.Dense(128,activation="relu")(batchnorm)
output_layer = tf.keras.layers.Dense(1,activation="relu")(dense3)
model = tf.keras.models.Model(input_layer,output_layer)
```

```python
model.summary()
```

```python
help(tf.keras.layers.Dropout)
```

```python
model.compile(optimizer = tf.keras.optimizers.RMSprop(),
         loss=tf.keras.losses.MeanSquaredError(),
         metrics=tf.keras.metrics.MeanSquaredError())
```

```python
#This "model.fit" itself will do the Backpropagation

model.fit(x_data,y_data,validation_split=0.2,epochs=200)

#Here we get Validation Loss and Train Loss by comparing these 2 we identify whether it is "bias/variance/" problem then we need to solve or else if it is optimal leave it
#We need to make suere that our "Bias&Variance" problems should not be occured.We need to get optimal loss or merely equal with less loss in both(Train&Test) 
```
