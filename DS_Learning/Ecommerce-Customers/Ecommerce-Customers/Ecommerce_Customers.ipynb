#### NOTE:In statistical analysis When we are going to build a ML Model After "EDA/Feature Engineering" we have to do statistical analysis,statistical analysis involves mostly "correlations".We have discussed that Inorder to find the "correlations" between each pair we have particular formulae for each combination based on its category type of the attributes.In Above after doing "EDA" we got the final dataset excluding "email,avatar and address" from dataset we got all "Numerical Attributes",while doing numerical to numerical we have one famous correlation formulae in "scipy.stats" that is called " pearsonr".pearsonr is a statistical formulae which is used to find the "correlation score" using this we can perform operations

```python
# Linear Regression Model Assumption is the Independent variable vector or Feature Vector has a property that each variable in feature vector is independent from other features

# hence we need to find whether there is any correlation between the variables of feature vector

# for that we need to calculate the Coefficent of Correlation between each pair of attributes .

# there are 5 Attributes, each attribute can have 4 other combinations so total we have  5*4=20 combinations

# for each combination we need to calculate coefficient of correlation (R)
from scipy.stats import pearsonr
print("Correlation with Label:")
for x in ecommerce.columns:
    y = "Yearly_Amount_Spent"
    if x!= y:
      print("Coefficient of Correlation R("+x+","+y+")  is:    "+str(pearsonr(ecommerce[x],ecommerce[y])[0])+"\n")
print("\n\nCorrelation between Attributes:")
for x in ecommerce.columns:
    for y in ecommerce.columns:
        if x!= y:
          print("Coefficient of Correlation R("+x+","+y+")  is:    "+str(pearsonr(ecommerce[x],ecommerce[y])[0])+"\n")
```

```python
#As we are doing a Experiment here we are doing some stuff which is not concerned.So,we can do whatever we want like here we separated "notrelated" and "related" So,we can just inspect/see the data here

# if the coefficient of correlation of (x,y) is approximately equal to 0 then those 2 features are independent

# if the coefficient of correlation of (x,y) is greater than 0.2 then those two have a relationship
notrelated=[] # It stores all "notrelated" pairs
related=[]    # It stores all "related" pairs
for x in ecommerce.columns:
  for y in ecommerce.columns:
    if x!=y:
      if pearsonr(ecommerce[x],ecommerce[y])[0] <= 0.2:
        notrelated.append((x,y))
      else:
        related.append((x,y))
print("Not Related pairs:")
for x in notrelated:
   print(x)
print("\n\nRelated pairs:")
for x in related:
  print(x)
```

```python
#Here "scikitlearn/sklearn" is a library inorder to split into train and test data
#Here we can also split manually instead we already have a function So,we can do using it 
#But we have a function "train_test_split()" in library "sklearn" using this function we can split
"""
"train_test_split()" function splits as 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25)" in this "test_size" that means
we have total number of examples are 500 in this how much percentage of number will be given to
test-set and how much percentage will be given to train-set based on this we should decide.Here we 
taken "test_size=0.25" it means 25% of training dataset given to test-set and remaining 75% given to
train-set we are asking to "train_test_split" function.

Here "train_test_split(X,Y,test_size=0.25)" in this "X" is a datframe and "Y" is a series not a 
dataframe.When we have given these the function will shuffle the whole data(X & Y) then it will 
pick Randomly 75% of the Data and put it in the training-set and Randomly put the take the "test_size"
of the inputs and put it in the set-block/test-bucket that's how it works

"""
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25)
```

```python
#This "LinearRegression" is already implemented in the Pseudo-code which we have seen already
# Here  "lr=LinearRegression()" we are just instantiating that particular class object "LinearRegression()" to 'lr'
# In this class object "LinearRegression()" fit is a function which we giving "X_train" and "Y_train"
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)

# Linear Regression best fit line will be returned i.e the coefficientf of a line will be returned 
# The line is Y=m1*Avg_Session_Length + m2*Time_on_App + m3*Time_on_Website + m4*Length_of_Membership + intercept

# (m1,m2,m3,m4) are coefficients and one intercept
```

```python
"""
Coefficients/Weights:
---------------------
In linear regression, "W" represents the weights or coefficients associated with 
each feature or independent variable. If there is only one feature or independent variable, 
then "W" will be a single value. However, if there are multiple features or independent variables, 
then "W" will be a vector of weights or coefficients, with each element of the vector corresponding 
to the weight or coefficient of a specific feature.

Intercept/bias:
---------------
In linear regression,"b" represents the bias term or the intercept. It is a single value that is 
added to the weighted sum of the input features to make the final prediction. Therefore, "b" is a 
single value in linear regression.
"""
```

```python
print("Coefficients:"+str(list(lr.coef_)))
print("Intercept:"+str(lr.intercept_))


#shape of W == (1,n) = (1,4)
#For these particular values w1,w2,w3,w4 and particular intercept value 'c' our Linear Function is 
#going to best fit I mean its getting the best values.So,these are the values of coefficient
#Here "Length_of_Membership" is a feature which has high weightage "61.81313127" compare to remaining
#Here "25.61617035, 38.3685783 ,  0.23021731, 61.81313127" are the values of "Weights" which is obtained at "Global Minimum"
```

```python
#Here it will predict "X_test" where "X_test" is a new dataset which is not trained
#Here the "predict()" function will predict "Y^/Y(pred)" for this new dataset
predict=lr.predict(X_test)
```

```python
#Here we are comparing the predicted values(Y^ or predict) and actual values(Y_test or Y) 
from matplotlib import pyplot as plt
plt.scatter(Y_test,predict)
plt.xlabel('Y_test')
plt.ylabel('predict')
```

```python
# Here "mean_squared_error()" is a function in "metrics" it implements "J(W,b)" convex function.which is used to caluclate "MSE" B/W predict,Y_test
# Here "MSE" is "87.97692113683021" for unknown dataset,which is really poor performance.We can also reduce this "MSE" even more
# But for now it is "87.97692113683021" 
# This value "87.97692113683021" is a minimum value of the whole function.we discussed that there 
# is a convex function that "convex function" minimum squared value(MSE) here "87.97692113683021" 
# this is also known as global minima sometimes 
# Here the "feature scaling" is also important.When we have our "Values of features are large numbers" it will impact on "Cost"
# Here We got MSE which is "Loss" 96.This "feature scaling" is also one of the factor which impacts or Model Performance 
from sklearn import metrics
print("Mean squared Error:"+str(metrics.mean_squared_error(predict,Y_test)))
```
