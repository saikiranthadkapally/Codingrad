```python
#Let's take an example
'''
Actual Image --- By default it is "RGB"
(200,988,109) --- 
R = 200
G = 98
B = 109

NOTE: Whenever the reddish color of images are turn into bluish color then we need to understand that the reading method of
a particular library is reading the values in such a way that it reading in reverse format as "BGR".Bydefault the opencv 
read the image in "BGR" format.

In point of view of "cv2" it will be seen as "RGB" as "BGR" --- Here the values will be reverse or inverted.
R = 109
G = 98
B = 200 

'''
```

```python
'''
NOTE:By default the "openCV" library read the image in "BGR" format.We need to convert it into "RGB" again We use "cvtColor"
is a function inside the "cv2" it can be write as "variable_name = cv2.cvtColor(input_image,object of cv2)".Here 
"objects of cv2" it means in "cv2" we don't send string or numbers as a parameter rather parameters are itself the object of 
"cv2".Thes objects are in-built objects.Mostly we need to deal with the "cv2" objects only.We don't pass strings or numbers 
as a parameters It is an old approach.We mostly use "objects of cv2".We remeber these functions automatically when we are
doing the project in "cv2".
'''
#Let's take above example
#Here "cv2.COLOR_BGR2RGB" is an inbuilt attribute for "cv2" module.
convertedImg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```
